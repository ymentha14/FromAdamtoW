```python
import pandas as pd
df = pd.read_json("./log_images_results.json")
df_results = pd.json_normalize(df['results'])
df_best_param = pd.json_normalize(df['best_param'])
df = pd.concat([df, df_best_param, df_results], axis=1)
df = df.drop(columns=['results', 'best_param'])
```

```python
df
```

### Adam

```python
adam = df[df['optim'] == 'Adam']
ax = pd.Series(adam.iloc[0]['train_accuracy']).plot.line()
ax.set_title("Train accuracy Adam");
ax;
```

```python
adam = df[df['optim'] == 'Adam']
ax = pd.Series(adam.iloc[0]['train_loss']).plot.line()
ax.set_title("Train loss Adam");
ax;
```

```python
adam = df[df['optim'] == 'Adam']
ax = pd.Series(adam.iloc[0]['val_loss']).plot.line()
ax.set_title("Val loss Adam");
ax;
```

```python
adam = df[df['optim'] == 'Adam']
ax = pd.Series(adam.iloc[0]['val_accuracy']).plot.line()
ax.set_title("Val acc Adam");
ax;
```
